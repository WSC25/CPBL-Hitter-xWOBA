```python
result_to_remove = ['Error','FoulBall','StolenBase','CaughtStealing']
data = data[~data['PlayResult'].isin(result_to_remove)]
```

```python
def calculate_so_go_count(series):
    sf_count = (series == 'Sacrifice').sum()
    hr_count = (series == 'HomeRun').sum()
    _3b_count = (series == 'Triple').sum()
    _2b_count = (series == 'Double').sum()
    _1b_count = (series == 'Single').sum()
    out_count = (series == 'Out').sum()
    E_count = (series == 'Error').sum()
    FC_count = (series == 'FieldersChoice').sum()
    return pd.Series({'HomeRun':hr_count,
                      'Sacrifice': sf_count,
                      'Triple': _3b_count,
                      'Double': _2b_count,
                      'Single': _1b_count,
                      'Out': out_count,
                     'Error':E_count,
                     'FieldersChoice':FC_count})
```

```python
result_counts = data.groupby(['ExitSpeed', 'Angle','Distance'])['PlayResult'].apply(calculate_so_go_count).reset_index()
result_counts = result_counts.rename(columns={'level_3': '打席結果', 'PlayResult': '次數'})
result_counts
```

```python
result_counts['次數'] = result_counts['次數'].astype(int)

final_resault = pd.pivot_table(result_counts, index=['ExitSpeed', 'Angle','Distance'], columns='打席結果', values='次數', aggfunc='sum', fill_value=0)

final_resault
```

```python
final_resault['安打']=final_resault['Double']+final_resault['Single']+final_resault['Triple']+final_resault['HomeRun']
final_resault['長打']=final_resault['Double']+final_resault['Triple']+final_resault['HomeRun']
final_resault['出局機率']=final_resault['Out']+final_resault['Sacrifice']+final_resault['Error']/final_resault['Out']+final_resault['Sacrifice']+final_resault['Error']+final_resault['Double']+final_resault['Single']+final_resault['Triple']+final_resault['HomeRun']
final_resault['一安機率']=final_resault['Single']/final_resault['安打']+final_resault['Out']+final_resault['Sacrifice']+final_resault['Error']
final_resault['二安機率']=final_resault['Double']/final_resault['安打']+final_resault['Out']+final_resault['Sacrifice']+final_resault['Error']
final_resault['三安機率']=final_resault['Triple']/final_resault['安打']+final_resault['Out']+final_resault['Sacrifice']+final_resault['Error']
final_resault['全壘打機率']=final_resault['HomeRun']/final_resault['安打']+final_resault['Out']+final_resault['Sacrifice']+final_resault['Error']


#final_resault['安打機率']=final_resault['安打']/final_resault['安打']+final_resault['Out']+final_resault['Sacrifice']+final_resault['Error']
#final_resault['長打機率']=final_resault['長打']/final_resault['安打']+final_resault['Out']+final_resault['Sacrifice']+final_resault['Error']
#final_resault['出局機率']=final_resault['出局']/final_resault['Double']+final_resault['Single']+final_resault['Triple']+final_resault['HomeRun']+final_resault['出局']
```

### LinearRegression

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score
```

```python
X = final_resault[['ExitSpeed(KM)', 'Angle']]
#,'Double','HomeRun','Out','Sacrifice','Single','Triple']]
y = final_resault[['安打機率', '長打機率','出局機率','Distance']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=40)

model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

# 打印预测结果
print(predictions)
```

```python
print('accuracy:', model.score(X_test, y_test))
```

```python
def predict_outcome_probability(ExitSpeed, Angle):
    prediction = model.predict([[ExitSpeed, Angle]])
    
    probability = {
        '安打機率': prediction[0][0],
        '長打機率': prediction[0][1],
        '出局機率': prediction[0][2],
        'Distance': prediction[0][3],
    }
    
    return probability

ExitSpeed = 110
Angle = 0
result = predict_outcome_probability(ExitSpeed, Angle)
print(result)
```
